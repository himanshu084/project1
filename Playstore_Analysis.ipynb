```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("googleplaystore.csv")
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace = True)
```

```python
df.Reviews=pd.to_numeric(df.Reviews,errors = "coerce") 
```

```python
df.Size.unique()
```

```python
def func(x):
    if('M' in x):
        return float(x[:-1])*(1024);
    if('k' in x):
        return float(x[:-1])
df.Size.apply(func)
```

```python
df.Size = df["Size"].apply(func)
```

```python
df
```

```python
df.Size.isna().sum()
```

```python
df.Installs.isna().sum()
```

```python
df
```

```python
def checkInstalls(x):
    if '+' in x:
        return x[:-1]
```

```python
df.Installs = df.Installs.apply(checkInstalls)
```

```python
def removeCommas(x):
    if(x==None):
        return None
    return x.replace(',','')
```

```python
df.Installs = df.Installs.apply(removeCommas)
```

```python
df
```

```python
df.Installs.isna().sum()
```

```python
def changePrice(x):
    if(x=='0'):
        return float(0)
    if('$' in x):
        return float(x[1:])
    return
    
```

```python
df.Price = df.Price.apply(changePrice)
```

```python
df
```

```python
df.Price.isna().sum()
```

```python
def clean(x):
    if(x=="Varies with device"):
        return np.nan
    if(type(x)==str):
        return x.replace(" and up","")
```

```python
df["Android Ver"] = df["Android Ver"].apply(clean)
```

```python
df.isna().sum()/df.shape[0]
```

```python
df.drop(index=10472,inplace=True)
```

```python
df["Android Ver"].fillna(df["Android Ver"].mode()[0],inplace=True)
```

```python
df.Rating.fillna(df.Rating.mean(),inplace = True)
```

```python
df.Size.fillna(df.Size.median(),inplace = True)
```

```python
df.dropna(inplace = True)
```
